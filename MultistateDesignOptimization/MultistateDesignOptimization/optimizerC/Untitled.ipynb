```python
import pylab
%matplotlib inline
import numpy as np

populationSize = [32,64,128,256,512]
simulationTime1 = [4.4508, 5.60139,10.289, 23.434,43.5072]
simulationTime2 = [5.34427, 7.46256, 10.2169, 16.3685, 26.9397]
simulationTime4 = [5.25715, 6.27206, 8.00413, 10.8053, 18.3358]

#1024 iterations. 
pylab.figure(figsize=[8,5])
pylab.plot(np.array(populationSize),np.array(simulationTime1),label='Single Thread')
#pylab.plot(np.array(populationSize),np.array(simulationTime2),label='Two Threads')
pylab.plot(np.array(populationSize),np.array(simulationTime4),label='Four Threads')
pylab.plot(np.array(populationSize), simulationTime1/np.array(4),label='Ideal: Single Thread/4',linestyle='dashed',color='blue')

pylab.xlabel('Population Size (1028 iterations)')
pylab.ylabel('Total Time (s)')
pylab.rcParams.update({'font.size': 14})
pylab.legend(fontsize=14,loc=2)

#1024 iterations. 
pylab.figure(figsize=[8,5])
pylab.plot(np.array(populationSize),np.array(simulationTime1)/(4*np.array(simulationTime4)),label='4 Thread')
pylab.plot(np.array(populationSize),np.array(simulationTime1)/(2*np.array(simulationTime2)),label='2 Thread')

pylab.xlabel('Population Size, 1028 iterations')
pylab.ylabel('Strong Scaling Efficiency')
pylab.rcParams.update({'font.size': 14})
pylab.legend(fontsize=14,loc=2)

#1024 iterations. 
pylab.figure(figsize=[8,5])
pylab.plot(np.array(populationSize),np.array(simulationTime1)/(np.array(simulationTime4)),label='Single Thread')
pylab.xlabel('Population Size, 1028 iterations')
pylab.ylabel('Weak Scaling Efficiency')
pylab.rcParams.update({'font.size': 14})
pylab.legend(fontsize=14,loc=2)
```
